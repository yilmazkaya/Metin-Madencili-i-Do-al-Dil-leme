```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
import nltk

# Uncomment the line below if you haven't downloaded the 'punkt' resource
# nltk.download('punkt')

# Read text data from the file
filename = "mevlana.txt"
with open(filename, 'r', encoding='utf-8') as file:
    text_data = file.read().splitlines()

# Tokenize the text data
tokenized_text_data = [' '.join(word_tokenize(doc)) for doc in text_data]

#print(tokenized_text_data)
# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(tokenized_text_data)
#print(tfidf_matrix)
# Example new documents



new_documents = [
    "Duydum ki bizi brakmaya azmediyorsun, etme.",
    "Baka bir yar, baka bir dosta meylediyorsun, etme.",
    "Sen yadeller dnyasnda ne aryorsun yabanc?"
]




# Transform the new documents using the vectorizer
new_documents_tfidf = vectorizer.transform(new_documents)

# Print vocabulary
print("\nVocabulary:")
print(vectorizer.vocabulary_)

print(new_documents_tfidf)
#print(new_documents_tfidf.toarray())


```
