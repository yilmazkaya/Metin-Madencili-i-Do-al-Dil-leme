```python
from transformers import VitsModel, AutoTokenizer #Hugging Face'in VITSTTS (Vision-Text-Speech) modelini
import torch
import scipy
from scipy.io.wavfile import write


model = VitsModel.from_pretrained("facebook/mms-tts-tur")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-tur")

metin = "Merhaba arkadaşlar Hugging Face uygulamalarına hoş geldiniz"
girdiler = tokenizer(metin, return_tensors="pt")

with torch.no_grad():
    output = model(**girdiler).waveform

# Varsayılan olarak 'output' bir PyTorch tensor olarak kabul edilir
# Bu nedenle onu bir NumPy dizisine dönüştürmemiz gerekiyor

from IPython.display import Audio

# Audio nesnesini kullanarak sesi oynat
Audio(output, rate=model.config.sampling_rate)


```
