# Word dosyalarından Okuma

```python
from docx import Document

def read_word_document(file_path):
    doc = Document(file_path)
    content = []

    for para in doc.paragraphs:
        content.append(para.text)

    return '\n'.join(content)

# Word belgesi okuma
filename = "mevlana.docx"
content = read_word_document(filename)

# Okunan metni ekrana yazdırma
print(content)
```

# Pdf dosylarından Okuma

```python
import PyPDF2

def read_pdf(file_path):
    text = ""
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text

# PDF dosyası okuma
pdf_filename = "mevlana.pdf"
content = read_pdf(pdf_filename)

# Okunan metni ekrana yazdırma
print(content)
```

# Excel dosyalarından okuma

```python
import pandas as pd

def read_excel(file_path, sheet_name):
        # Excel dosyasını oku
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        # Okunan veriyi ekrana yazdır
        print(df)

        # DataFrame'i döndür
        return df
   


excel_filename = "mevlana.xlsx"

data_frame = read_excel(excel_filename, "Sayfa1")
```

# HTML dosyasından okuma

```python
import chardet

def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        result = chardet.detect(file.read())
        return result['encoding']

# Dosyanın karakter kodlamasını belirleme
encoding = detect_encoding("mevlana.htm")
print(f"Dosyanın karakter kodlaması: {encoding}")

# Dosyayı bu karakter kodlamasıyla açma
with open("mevlana.htm", 'r', encoding=encoding) as file:
    html_content = file.read()
    print(html_content)

```
