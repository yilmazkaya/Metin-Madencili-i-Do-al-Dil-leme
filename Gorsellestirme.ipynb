```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# CSV dosyasn oku
filename = "WeatherReports.csv"
tbl = pd.read_csv(filename, dtype={'YORUM': str})
tbl.iloc[:, 1] = tbl.iloc[:, 1].astype(str)

textData = tbl.iloc[:, 1]
# Metin verilerini al
print(textData.head(10))

# Kelime bulutu olutur
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(textData))

# Kelime bulutunu gster
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Hava Durumu Raporlar")
plt.show()

```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# CSV dosyasn oku
filename = "WeatherReports.csv"
tbl = pd.read_csv(filename, dtype={'YORUM': str, 'event_type': str})
tbl.iloc[:, 1] = tbl.iloc[:, 1].astype(str)

textData = tbl.iloc[:, 1]
# Metin verilerini al
textData = tbl['YORUM']
labels = tbl['SINIF']


plt.figure(figsize=(20, 10))
# Hail etkinlii iin kelime bulutu olutur
plt.subplot(1, 2, 1)
hail_idx = labels == "Hail"
wordcloud_hail = WordCloud(width=1200, height=600, background_color='white').generate(' '.join(textData[hail_idx]))
plt.imshow(wordcloud_hail, interpolation='bilinear')
plt.axis('off')
plt.title("Hail")

# Heavy Rain etkinlii iin kelime bulutu olutur
plt.subplot(1, 2, 2)
heavy_rain_idx = labels == "Heavy Rain"
wordcloud_heavy_rain = WordCloud(width=1200, height=600, background_color='white').generate(' '.join(textData[heavy_rain_idx]))
plt.imshow(wordcloud_heavy_rain, interpolation='bilinear')
plt.axis('off')
plt.title("Heavy Rain")

# Grseli gster
plt.show()
```
