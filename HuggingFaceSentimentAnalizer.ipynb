```python
import pandas as pd

# Duyguanalizi.csv dosyasını oku
df = pd.read_csv('duyguanalizi.csv',encoding="iso-8859-1")


from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

# Duygu analizi modelini yükleyelim
model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
sa= pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)



# Veri çerçevesindeki her bir yorumu sınıflandıralım
for index, row in df.iterrows():
    comment = row['Yorum']
    result = sa(comment)[0]
    
    # Duygu sınıfını ve skorunu ekrana yazdıralım
    print(f"Yorum: {comment}")
    print(f"Duygu Sınıfı: {result['label']}, Skor: {result['score']:.4f}")
    print()
```
