EŞ OLUŞUM AĞLARI
Doğal dil işlemede eş oluşum ağları (co-occurrence networks) yaygın olarak kullanılan bir yöntemdir. Eş oluşum, bir metinde belirli kelimelerin veya terimlerin birlikte ne sıklıkta geçtiğini ifade eder. Eş oluşum ağları, bu terimlerin birbiriyle olan ilişkilerini görselleştirmek ve analiz etmek için kullanılır.

Eş oluşum ağları genellikle aşağıdaki adımları içerir:

1-Metni Tokenlara Ayırma: Metni kelimelere veya terimlere ayırma işlemi yapılır.

2-Eş Oluşum Matrisi Oluşturma: Her kelimenin diğer kelimelerle olan eş oluşumlarını sayan bir matris oluşturulur.

3-Eş Oluşum Ağını Oluşturma: Eş oluşum matrisi temel alınarak bir ağ (graph) oluşturulur. Kelimeler düğüm (node) olarak, eş oluşumlar ise kenarlar (edge) olarak temsil edilir.

4-Ağ Analizi ve Görselleştirme: Oluşturulan ağ üzerinde çeşitli analizler yapılır ve ağı görselleştirmek için grafik araçları kullanılır.

Bu tür ağlar, kelime benzerliği, metin sınıflandırma, öneri sistemleri ve diğer doğal dil işleme görevlerinde kullanılabilir. Eş oluşum ağları, kelime ilişkilerini anlamak ve metin verilerini keşfetmek için güçlü bir araçtır.

```python
import nltk
import networkx as nx
import matplotlib.pyplot as plt
from nltk import FreqDist
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Metni örnek bir metinle değiştirin
text = "Doğal dil işleme, bilgisayarların insan dilini anlamasına ve yorumlamasına odaklanan bir alandır. insan dilini anlama. Merhaba bugün hava çok güzel. Doğal dil işleme. Bilgisyarını unutma"
        
# Metni kelimelere ayır
tokens = word_tokenize(text)


tokens = [word.lower() for word in tokens]

# Eş oluşumları say
co_occurrence_matrix = nltk.FreqDist(nltk.bigrams(tokens))

# Eş oluşum ağını oluştur
co_occurrence_network = nx.Graph()
for bigram, freq in co_occurrence_matrix.items():
    co_occurrence_network.add_edge(bigram[0], bigram[1], weight=freq)

# Görselleştirme
pos = nx.spring_layout(co_occurrence_network)
nx.draw(co_occurrence_network, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_color='black', font_size=8, edge_color='gray')
labels = nx.get_edge_attributes(co_occurrence_network, 'weight')
nx.draw_networkx_edge_labels(co_occurrence_network, pos, edge_labels=labels)

# Görseli göster
plt.show()
```

```python
import nltk
import networkx as nx
import matplotlib.pyplot as plt
from nltk import FreqDist
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import shakespeare

# Read the text from the file
filename = "mevlanatek.txt"
with open(filename, "r", encoding="utf-8") as file:
    text = file.read()
    


# Metni kelimelere ayır
tokens = word_tokenize(text)


# Stopwords'ları ve noktalama işaretlerini temizle
filtered_tokens = [word.lower() for word in tokens]

# Eş oluşumları say
co_occurrence_matrix = nltk.FreqDist(nltk.bigrams(filtered_tokens))

# Eş oluşum ağını oluştur
co_occurrence_network = nx.Graph()
for bigram, freq in co_occurrence_matrix.items():
    co_occurrence_network.add_edge(bigram[0], bigram[1], weight=freq)

# Görselleştirme
pos = nx.spring_layout(co_occurrence_network, seed=42)  # seed kullanarak aynı yerleşimi elde etmek için
plt.figure(figsize=(12, 8))
nx.draw(co_occurrence_network, pos, with_labels=True, font_weight='bold', node_size=800, node_color='skyblue', font_color='black', font_size=8, edge_color='gray')
labels = nx.get_edge_attributes(co_occurrence_network, 'weight')
nx.draw_networkx_edge_labels(co_occurrence_network, pos, edge_labels=labels)

# Görseli göster
plt.title("Mevlana")
plt.show()
```
